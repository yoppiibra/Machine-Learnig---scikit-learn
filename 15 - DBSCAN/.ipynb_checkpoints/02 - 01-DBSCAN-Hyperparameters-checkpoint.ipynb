## Label Discovery

```python
def display_categories(model, data):
    label = model.fit_predict(data)
    sns.scatterplot(x='X1', y='X2', data= data, palette='Set1', hue=label)
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
help(DBSCAN)
```

```python
model = DBSCAN()
```

```python
display_categories(model, two_blobs);
```
